```python
import csv

with open("ChinaCompleteDataset.csv") as csvfile:
    chinaFile = csv.reader(csvfile)   
    chinaWithBorders = [pair for pair in chinaFile][1:]
    chinaNames = list(set([x[0] for x in chinaWithBorders]))
    
#chinaWithBorders[:10]
#chinaNames[:10]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

CN = nx.Graph()
CN.add_nodes_from(chinaNames)
CN.add_edges_from(chinaWithBorders)
```

```python
nx.spring_layout(CN)

plt.figure(figsize=(22,22))

pos = nx.spring_layout(CN)
nx.draw_networkx_nodes(CN, pos, 
                        nodelist=CN.nodes(),
                        node_color="yellow",
                        alpha=0.4)
nx.draw_networkx_labels(CN, pos, 
                        nodelist=CN.nodes(), 
                        font_size=18)
nx.draw_networkx_edges(CN, pos, 
                        edgelist=CN.edges(),
                        edge_color="grey")
plt.axis('off')
plt.show()
```
